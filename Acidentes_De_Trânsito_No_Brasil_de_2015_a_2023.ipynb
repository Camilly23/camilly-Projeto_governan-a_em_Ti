# Evolução

```python
import matplotlib.pyplot as plt

anos = [2019, 2020, 2021, 2022, 2023]
colisoes = [120, 100, 110, 90, 95]
atropelamentos = [50, 45, 55, 60, 58]
capotamentos = [20, 15, 18, 25, 23]

plt.plot(anos, colisoes, label='Colisões', marker='o')
plt.plot(anos, atropelamentos, label='Atropelamentos', marker='o')
plt.plot(anos, capotamentos, label='Capotamentos', marker='o')

plt.xlabel('Ano')
plt.ylabel('Número de Ocorrências')
plt.title('Evolução dos Acidentes de Trânsito')
plt.legend()
plt.grid(True)
plt.show()
```

O gráfico mostra a evolução dos números de ocorrências de colisões, atropelamentos e capotamentos ao longo dos anos. Cada linha no gráfico representa um tipo de acidente específico.

Com base nos dados fornecidos, podemos fazer as seguintes observações:

O número de colisões tem uma tendência geral de diminuição ao longo dos anos, com uma leve queda de 2019 a 2022 e um aumento discreto em 2023.
O número de atropelamentos mostra uma tendência estável com pequenas variações. Houve um aumento em 2021, seguido de uma leve queda em 2022 e um pequeno aumento em 2023.

O número de capotamentos tem uma tendência geral de aumento ao longo dos anos, com um aumento significativo de 2019 a 2021, seguido de uma diminuição em 2022 e um leve aumento em 2023.

Essas observações permitem ter uma ideia da evolução dos diferentes tipos de acidentes de trânsito ao longo do tempo. O gráfico ajuda a identificar possíveis tendências e variações nos números de ocorrências, fornecendo informações importantes para a análise e prevenção de acidentes de trânsito.






# Estatísticas 

```python
import numpy as np
import statistics

# Dados de acidentes de trânsito (substitua pelos seus dados reais)
acidentes = [120, 100, 90, 80, 110, 95, 105, 115]

# Média
media = np.mean(acidentes)
print("Média:", media)

# Moda
moda = statistics.mode(acidentes)
print("Moda:", moda)

# Mediana
mediana = np.median(acidentes)
print("Mediana:", mediana)
```

```python
import matplotlib.pyplot as plt

# Dados estatísticos
valores = [media, moda, mediana]
labels = ['Média', 'Moda', 'Mediana']

plt.bar(labels, valores)
plt.ylabel('Valor')
plt.title('Valores Estatísticos')

plt.show()
```

O gráfico apresenta os valores estatísticos (média, moda e mediana) dos dados de acidentes de trânsito. Cada barra representa um valor estatístico, sendo a barra "Média" correspondente à média dos acidentes, a barra "Moda" correspondente à moda dos acidentes e a barra "Mediana" correspondente à mediana dos acidentes. O eixo y representa o valor numérico dessas estatísticas. O gráfico fornece uma visualização clara e comparativa das medidas estatísticas dos acidentes de trânsito.

# Regressão Linear 
 

```python
import numpy as np
import matplotlib.pyplot as plt

# Dados simulados de anos e número de acidentes de trânsito
anos = np.array([2015, 2016, 2017, 2018, 2019, 2020, 2021])
acidentes = np.array([100, 120, 90, 80, 110, 95, 105])

# Realiza a regressão linear
coeficientes = np.polyfit(anos, acidentes, 1)
regressao = np.poly1d(coeficientes)

# Gera os valores previstos pela regressão
anos_pred = np.linspace(min(anos), max(anos), 100)
acidentes_pred = regressao(anos_pred)

# Plota o gráfico com os dados e a regressão linear
plt.scatter(anos, acidentes, color='blue', label='Dados')
plt.plot(anos_pred, acidentes_pred, color='red', label='Regressão Linear')
plt.xlabel('Ano')
plt.ylabel('Número de Acidentes')
plt.title('Regressão Linear de Acidentes de Trânsito')
plt.legend()

plt.show()
```
